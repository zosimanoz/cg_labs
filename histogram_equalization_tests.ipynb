```python
# Get intensity of pixels in the image

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("lena_gray.bmp", cv2.IMREAD_GRAYSCALE)
w, h = img.shape[:2]

frequency = {}

for row in range(w):
    for column in range(h):
        intensity = img[column][row]
        count = frequency.get(intensity, 0)
        frequency[intensity] = count + 1

print("Unique Intensities " + str(len(frequency)))

freq_sorted = sorted(frequency.items()) 
x, y = zip(*freq_sorted)
plt.plot(x, y)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("beans.png", cv2.IMREAD_GRAYSCALE)
w, h = img.shape[:2]

eq_img = equ = cv2.equalizeHist(img) 

hist = cv2.calcHist([eq_img], [0], None, [256], [0, 256])
plt.hist(eq_img.ravel(), 256, [0, 256])
plt.title("Histogram for gray scale picture")
plt.show()

plt.imshow(eq_img, cmap=plt.get_cmap('gray'))
plt.show()
```
